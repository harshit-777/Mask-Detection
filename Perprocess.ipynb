```
import cv2 
import os 


```

```
folder_path = '/content/drive/.shortcut-targets-by-id/1e2-Ms1e9ar79MMl7gtEzQ8fXKad5yFP0/Face Mask Detection/dataset'

#we have two categorires so assiging folder path

category = os.listdir(folder_path)

#we have two types of labels 

label = [i for i in range(len(category))]

#converting to dict
label_dict = dict(zip(category,label))

print(label)
print(category)
print(label_dict)
```

```
# using a default size of  100 
image_size = 100
data = list()
target = list()

for cat in category:
  path = os.path.join(folder_path,cat)
  image_name = os.listdir(path)
  #print(image_name)

  for name in image_name:
    img_path = os.path.join(path,name)
    img = cv2.imread(img_path)
    #print(img)

    #converting it to gray color
    gray_conv = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
    #resize it to our default 
    resize_img = cv2.resize(gray_conv,(image_size,image_size))

    #append above images in data list 
    data.append(resize_img)
    target.append(label_dict[cat])


```

```
import numpy as np

#we will normalize our data by dividing 255 to get value b/w 0 and 1

data = np.array(data)/255.0
data.shape
```

```
data = np.reshape(data,(data.shape[0],image_size,image_size,1))
data.shape
#we have converted to 4-D means (batch_size,height,width,color channel)
```

```
target = np.array(target)
print(target)
print(target.shape)
```

```
from keras.utils import np_utils
```

```
target_cat = np_utils.to_categorical(target)
print(target_cat)
print(target_cat.shape)
```
